```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

avgIncome = pd.read_excel('avgIncome.xlsx', thousands=',')
avgIncome
```

```python
avgIncome = avgIncome.head(17)
avgIncome
```

```python
avgIncome = avgIncome.rename(columns= {'년도' : 'year',
                                       '전체' : 'total',
                                       '남성' : 'male',
                                       '여성' : 'female',
                                       '월임금총액' : 'total_income'})
avgIncome
```

```python
avgIncome = avgIncome.drop(columns=['total_income', 'total'])
avgIncome
```

```python
avgIncome.dtypes
```

```python
avgIncome['male'] = avgIncome['male'].astype(int)
avgIncome['female'] = avgIncome['female'].astype(int)
avgIncome['year'] = avgIncome['year'].astype(int)
avgIncome.dtypes
```

```python
avgIncome_sex = pd.melt(avgIncome, id_vars=['year'], value_vars=['male', 'female'], var_name='sex', value_name='avg_income')
avgIncome_sex
```

```python
avgIncome_sex.describe()
```

```python
sns.lineplot(data=avgIncome_sex, x='year', y='avg_income', hue='gender');
```
