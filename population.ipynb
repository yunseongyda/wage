```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
population = pd.read_excel('population.xlsx', thousands=',')
population
```

```python
population = population.head(17)
population
```

```python
population = population.rename(columns= {'연도' : 'year',
                                        '남자' : 'male',
                                         '여자' : 'female'})

population
```

```python
population['year'] = population['year'].astype(int)
population['male'] = population['male'].astype(int)
population['female'] = population['female'].astype(int)
```

```python
population.dtypes
```

```python
population_sex = pd.melt(population, id_vars=['year'], value_vars=['male', 'female'], var_name='sex', value_name='population')

plt.figure(figsize=(10, 6))
sns.barplot(data= population_sex, x = 'year', y = 'population', hue='sex').set(ylim = [0, 30000], title = 'Population by Year and Gender (2006-2022)');
```
