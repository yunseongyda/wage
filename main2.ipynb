```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # 한글 폰트의 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
E_2013 = pd.read_excel('Employed_2013.xlsx')
E_2018 = pd.read_excel('Employed_2018.xlsx')
E_2023 = pd.read_excel('Employed_2023.xlsx')
```

```python
E_2013
E_2018
E_2023
```

```python
E_2013['성별'] = E_2013['성별(남자)'] + E_2013['성별(여자)']
E_2018['성별'] = E_2018['성별(남자)'] + E_2018['성별(여자)']
E_2023['성별'] = E_2023['성별(남자)'] + E_2023['성별(여자)']
```

```python
# 그래프 크기 설정
plt.figure(figsize=(10, 6))

sns.barplot(data=E_2013, x='성별', y='직업별', color='skyblue', label='남자')
sns.barplot(data=E_2013, x=-E_2013['성별(여자)'], y='직업별',  color='orange', label='여자')

plt.title('2013년 상위 월급');
```

```python
# 그래프 크기 설정
plt.figure(figsize=(10, 6))

sns.barplot(data=E_2018, x='성별', y='직업별', color='skyblue', label='남자')
sns.barplot(data=E_2018, x=-E_2018['성별(여자)'], y='직업별',  color='orange', label='여자')

plt.title('2018년 상위 월급');
```

```python
# 그래프 크기 설정
plt.figure(figsize=(10, 6))

sns.barplot(data=E_2023, x='성별', y='직업별', color='skyblue', label='남자')
sns.barplot(data=E_2023, x=-E_2023['성별(여자)'], y='직업별',  color='orange', label='여자')

plt.title('2023년 상위 월급');
```
